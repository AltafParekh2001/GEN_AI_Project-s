```python
import inspect
# ‚úÖ YouTube Video Transcript & Summarizer (Bug-Free Version)
!pip install --upgrade youtube-transcript-api transformers

from youtube_transcript_api import YouTubeTranscriptApi
from transformers import pipeline
import re

def get_video_id(url):
    """Extract YouTube video ID from URL."""
    pattern = r"(?:v=|/)([0-9A-Za-z_-]{11}).*"
    match = re.search(pattern, url)
    return match.group(1) if match else None


def get_youtube_transcript(url, lang='en'):
    video_id = get_video_id(url)

    try:
        # Create instance first
        ytt_api = YouTubeTranscriptApi()

        # Use .fetch() method
        fetched_transcript = ytt_api.fetch(video_id, languages=[lang, 'en'])

        # Convert to raw data
        transcript = fetched_transcript.to_raw_data()

        # Combine text
        text = " ".join([entry['text'] for entry in transcript])
        return text
    except Exception as e:
        print(f"Error: {e}")
        return None


def summarize_text(text):
    """Summarize long text using a transformer model."""
    summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
    summary = summarizer(text, max_length=200, min_length=50, do_sample=False)
    return summary[0]['summary_text']

# üé• Example usage
url = input("Enter YouTube video URL: ")
transcript = get_youtube_transcript(url)

if transcript:
    print("\nüìù Transcript Preview:\n")
    print(transcript[:1000], "...")

    print("\nüîç Generating Summary...\n")
    summary = summarize_text(transcript[:4000])
    print("\nüìò Summary:\n", summary)
```
